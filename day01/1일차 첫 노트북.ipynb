# 코랩 환경 이해하기
## 코드블록과 텍스트 블록
- 코드블록은 파이썬 코드를 작성하고 실행할 수 있는 블록으로 , ctrl + Enter 를 누르면 해당 코드블록을 실행할 수 있다.
-텍스트 블록은 실행할 수 있는 블록은 아니지만 설명을 적거나 여러 블록을 묶는 기능을 하는 블록이다.


### 코드 블록
- 기본실행은 `ctrl + Enter`
- 실행 후 아래쪽 블록으로 이동은 `Shift + Enter`을 누르면 된다.
실행한 결과는 해당 코드 블록 아래쪽에 접을 수 있는 형태로 출력된다.

### 텍스트 블록
-'마크다운'이라느누 문법을 이용해서 텍스트 문서의 내용에 서식을 적용할 수 있는 블록이다.
- 이러한 마크다운의 문법은 차근차근 써보면서 익숙해지도록 한다.
-혹은 텍스트 블록 상단에 있는 버튼은 활용해서 서식을 적용할 수도 있다.


```python
# 코드 블록을 실행하려면 ctrl + Enter를 누르면 된다.
1+3

#코드 블록을 실행하고 다음 블록으로 이동하려면
#Shift +Enter를 누르면 된다.
2 + 3
# 이때 아래 블록이 없으면 코드블록을 만들면서 이동한다.
```
