# 변수와 자료형
- 변수란, 변할 수 있는 값을 저장해두는 저장공간이다.
- 변화하는 값에 `변수명`이라는 변하지 않는 이름을 부여함으로써, 항상 같은 이름으로 해당 값에 접근(acess)할 수 있다.
-변수에 값을 저장하고 변수명을 부여함으로써 컴퓨터와 개발자가 같은 대상을 가리킬 이름을 정하는 것을 선언/정의라고 한다.
```python
# 변수를 만드는 방법
# 변수명 = 값

```


````python
age = 30
age # 코드 블록 가장 아래쪽에 적힌 값만 출력이 된다.
````

## 주석(Comment)
- 코드 블록 안에서만 작성할 수 있는 것으로, 컴퓨터가 코드를 실행할 때 무시할 부분을 나타낼 때 사용한다.
- 주석 작성은 # 을 작성한 뒤 내용을 작성하면 해당 내용은 코드를 실행할 때 무시한다.
### 주석의 목적
- 개발자 간에 소통을 하기 위해서 자연어로 설명을 적어두는 공간으로 사용된다.
- 해당 코드에 대해서 설명하거나 사용방법에 대해서 설명할 때 쓰인다.
- 특정 코드를 **주석 처리**하면 해당 코드를 실행하지 않게 된다.
  - 주석 처리된 코드는 실행되지 않기 때문에 실행을 안시적으로 막아두는 용도로 주석을 사용한다.

## 변수의 초기화와 사용
- 초기화(initialize)란, 변수를 선언하고 값을 저장하는 것을 통틀어 이르는 말이다.
```python

```

- 만약에 이미 존재하는 변수에 새로운 값을 저장(=)하면 기존 값은 사라지고 새로운 값이 저장된다. (덮어쓰기)
- 변수에 값을 **저장**하는 것은 **등호**(=) **왼쪽**에서만 해당하고, 등호 오른쪽 혹은 그 외의 공간에서는 해당 변수에 담긴 값을 호출(call)하여 **사용**한다.


````python
# 변수의 초기화
age = 30

# 변수에 새로운 값을 덮어쓰기
age = 35
age
````

````python
# 변수명 명명규칙
# 1. 변수명은 한글, 영어, 숫자 모두 가능
# 2. 숫자로 시작하는 변수명은 불가능하다.
# 3. 공백 문자를 사용할 수 없다. 예) 내 나이 = 30
# 4. 특수문자는 언더바(_)만 가능하다.
# 변수명 작성법
# 파이썬에서 변수명을 만들 때에는 서로 다른 단어 사이에 언더바(_)를 작성하여
# 단어와 단어를 구분할 수 있게 한다. -> 가독성
# programming_language_python
# -> 스네이크 표기법
my_age = 25
my_name = '문승신'
````

## 자료형(datatype)
- 자료형이란 값을 다루는 개발자 입장에서 서로 다른 형식을 가진 값들의 유형을 정해놓지 않으면 연산을 할 때 곤란해진다.
  - 숫자와 문자를 덧셈활 수 있는가? -> 못 함    
  - 문자형으로 작성된 숫자를 숫자와 연산하고자 한다면 둘
  모두 숫자형으로 바꿔야 정상적으로 연산이 되도록
  시스템이 만들어져 있다.
- 이러한 자료형은 크게 세가지 타입으로 나뉜다.
  1. 숫자형(number)
      약칭: num
    - 정수(integer)
        약칭: int
      - 소수점이 없는 숫자
      - 개수를 셀 때 쓰인다.
    - 실수(float)
        약칭: float
      - 소수점이 있는 숫자
  2. 문자형
    - 문자(charactor)
        약칭: char,chr
    - 문자열(String)
        약칭: str
        - 작은 따옴표' 혹은 큰 따옴표 한 쌍으로 값을 감싸면 해당 값은 문자열로 취급한다.
        - 예) "30" # 30이라는 값을 문자열로 취급
  3. 논리형
      - 참과 거짓 두 가지 값만 존재하는 자료형
      - `True` 와 `False`만 존재한다.

## 입력문(input)
- 코드에 직접 작성되어서 실행 후에도 변경할 수 없게 작성한 코드를 **하드 코딩(hard coding)**되어 있다고 표현한다.


## 형변환(type casting)
- 문자열로 작성된 숫자값은 산술적인 연산에 사용할 수가 없다.
- 때문에 숫자만으로 이루어진 문자열 값을 숫자로 바꿔주는 작업이 필요하다
- 이때 이렇게 유형을 변환하는 것을 **형 변환**이라고 부르며, 그 방법은 모두 동일하다.
```python
# 바꾸려는 타입
# 정수 int()
# 실수 float()
# 문자형 str()
# 논리형 bool()
age = int("30")
```

````python
# 바꾸려는 타입
# 정수 int()
# 실수 float()
# 문자형 str()
# 논리형 bool()
age = int("40")
````

````python
# input()으로 입력되는 값은 모두 문자열이다
# 이렇게 입력된 문자열 값을 숫자형으로 바꿔서 연산하고자 할 때에는
# 숫자형으로 바꿔주어야 한다.
age = int(input("나이를 입력해주세요: "))

# 해당 값이나 변수에 담긴 값의 자료형을 검사하고자 할 때에는
# type()함수에 검사할 값을 넣어주면 된다.
age_type = type(age) # 변수 age의 타입을 변수 age_type에 저장한다.
print(age_type) # 변수 age_type에 담긴 값을 print해라
````

````python
# 문자열str을 실수형 float로 변환
# float()
# input_var = # 1. input()함수를 사용하여 문자열 입력을 받아
# float()함수를 사용하여 실수형으로 변환한 뒤
# 변수 input_var에 저장해라
input_str = input("실수 값을 입력해주세요: ")
input_var = float(input_str)
# 2. 변수 input_var를 print()함수에 넣어서 출력해라
print(input_var)

# 문자열str을 논리형 bool로 변환
# bool()
# 1. input() 함수를 이용하여 입력된 문자열값을
# bool 타입으로 변환하여 변수 input_bool에 저장해라
input_bool = bool(input("아무 말이나 입력해주세요: "))
# 2. 변수 input_bool을 print()함수에 넣어서 출력해라
print(input_bool)
# 논리값으로 형변환할 때에는 빈문자열"", 0, None 등은 False로 변환된다.
print(bool("")) # alt + shift + 위아래방향키 -> 한줄 복사
print(bool(0)) # False
print(bool(None)) # False
print(bool("0")) # True
````

## 출력문 심화
- 기본적으로 print()함수에 넣어주고 싶은 경우, 반점을 이용하여 여러 값을 한번에 출력할 수 있다.
- 이때 전달된 각 값 사이에 공백문자가 추가되어 출력된다.
```python
print("홍길동:", 30, "살") # 홍길동: 30 살
```
### f-string으로 문자열 구성하기
- f-string이란, 문자열 앞에 f를 붙임으로써 변수나 값의 연산식을 문자열 중간에 삽입할 수 있는 방법이다.
```python
name = "홍길동"
age = 30
print(f"{name}:{age}살") # 홍길동: 30살
```


````python
print("홍길동:", 30, "살") # 홍길동: 30 살
````

````python
# f-string으로 내용 출력하기
name = "홍길동"
age = 30
print(f"{name}: {age}살") # 홍길동: 30살

cal_age = f"{age + 10 }" #기존 30이라는 숫자값에 10을 더한 뒤 문자열로 변환
print(type(cal_age)) # str
print(cal_age) # 40
````

````python
#print()함수에서 반점을 사용했을 때 공백문자를 제거하는 방법
# sep 매개변수의 값을 설정해준다.
# 구분자(seperator)를 설정해주는 sep매개변수를 특정 문자열로 설정하면
# 반점이 등장한 위치에 해당 문자열이 삽입되어 각 값을 구분해준다.
print("홍길동:", 30, "살", sep=" ")
````
