```python
# part2_method_practice.py
# 함수 사용 연습하기
# 덧셈, 뺄셈, 곱셈, 나눗셈을 함수를 이용하여 구현해보자.

# num1, num2은 매개변수이자 연산자로 보자면 피연산자에 속한다

# 덧셈 함수
def add(num1, num2):
    result = None
    # 비즈니스 로직
    result = num1 + num2
    return result

# 뺄셈 함수
def sub(num1, num2):
    result = None
    # 비즈니스 로직
    result = num1 - num2
    return result

# 곱셈 함수
def multi(num1, num2):
    result = None
    # 비즈니스 로직
    result = num1 * num2
    return result

# 나눗셈 함수
def div(num1, num2):
    result = None
    # 비즈니스 로직
    result = num1 / num2
    return result

# 매개변수명만 소괄호 안에 있으면 반드시 값을 받아야 하는 매개변수라는 의미이다.
# 이때 전달받는 인자, 인수값은 순서에 따라서 자동으로 배정되기도 하지만(positional argument)
# 정확한 매개변수에 원하는 값을 넣고자 한다면
# 함수를 호출, 사용하는 쪽에서 매개변수명=값 과 같이
# 직접 값을 할당할 수도 있다.
# def 함수명(매개변수1, 매개변수2):
#     pass

def sub2(num1, num2):
    result = num1 - num2
    return result

if __name__ == "__main__":
    # print(add(1, 3)) # 4
    # print(sub(10, 3)) # 7
    # print(multi(3, 3)) # 9
    # print(div(10, 3)) # 3.333333...

    # 함수를 사용할 때 매개변수의 이름을 정확하게 작성하여 값을 할당하면
    # 순서와 무관하게 값을 전달할 수 있다.
    # 이때, 순서대로 값을 넣는 방식을 positional argument라고 하며
    # 매개변수명에 직접 값을 할당하는 방식은 반드시 순서방식 이후에
    # 사용해야 한다.
    result = sub2(num2=10, num1=7)
    result = sub2(10, num2=7)
    print(result) # -3
```
