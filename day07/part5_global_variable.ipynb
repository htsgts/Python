```python
# part5_global_variable.py
# 전역 변수와 지역 변수
# 전역변수(global variable)
# 전역 변수란, 함수나 제어문 안에서 선언되지 않은,
# 들여쓰기 없이 작성된 변수를 가리킨다.
# 이러한 전역변수는 어떤 함수나 제어문 안에서 사용되고 호출될 수 있다.
file_path = "C:/file.py" # 전역변수
def test():
    # 전역변수 호출
    print(file_path)

# 지역변수(local variable)
# 지역변수는 함수 혹은 제어문 안에서 처음 만들어진 변수를 가리킨다.
# 이러한 지역변수는 일반적으로 해당 영역을 벗어나는 순간
# 메모리에서 해제되어 호출할 수 없는 상태가 된다.(사라진다.)

def test_local_var():
    # 지역변수 선언
    local_var = "지역변수값"
    print(local_var)

# 만약에 함수 안에서 전역변수를 선언하고자 한다면
# 앞에 global 예약어를 작성하여 전역변수임을 명시할 수 있다.
def test_global_var():
    global local_var # 함수 내부에서 전역변수를 선언, 정의
    # 이 함수가 실행되어야 전역변수를 할당하게 된다.
    local_var = "함수에서 값을 할당한 전역변수"

if __name__ == "__main__":
    test()
    # 전역변수 호출
    # print(file_path)
    # test_local_var() # 지역변수 테스트
    test_global_var() # 함수에서 전역변수 선언
    print(local_var)

    # 전역변수를 사용하는 경우
    # 해당 .py파일 안에서 공통적으로 사용되는 값을 저장할 때
    # 전역변수를 활용한다.
    # 지역변수를 사용하는 경우
    # 함수 내부에서만 임시적으로 사용하고 외부에서
    # 굳이 접근할 필요가 없는 변수에 대해서는 지역변수를 활용한다.
    # -> 웬만하면 지역변수, 특수한 경우엔 전역변수
```
