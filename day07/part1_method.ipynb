```python
# part1_method.py
# 메서드, 함수(function)란?
# 함수는 미리 정의한 순서대로 코드가 동작하도록 만들어둔
# 잘 돌아가는 부품이다.
# 함수는 전달되는 매개변수를 정하고
# 그 매개변수에 담긴 값을 조작, 처리하여
# 그 결과 return하는 동작을 수행하는, 반복적으로 사용 가능한 기성품 코드 블록이다.

# ========
# 임포트 공간
# ========
# 비즈니스 로직
# 함수는 비즈니스 로직 공간에 작성한다.
# 함수의 시작은 def로 시작
# def 함수명(매개변수):
def print_name(): # 입력받는 매개변수가 없다.
    # 외부로부터 입력받아서 사용하는 값을 다루지 않는다.
    # 실제 로직이 구현되는 부분
    print("이준상")

    # return 예약어를 작성한 뒤 값을 뒤에 작성하면
    # 함수를 사용했던 위치와 해당 값이 치환되게 된다.
    return

# 매개변수는 있는데, 반환값은 없는 함수 만들기
def print_text(text):
    # 함수 바깥에서 전달받은 값을 print()함수에 사용하여 출력하는 기능
    # 전달된 값에 다른 조작을 가해서 내부에서 내용을 변경할 수 있다.
    text = text + "함수 내부에서 추가된 값"
    print(text)
    return

# 매개변수도 있고, 반환값도 있는 함수 만들기
def make_full_name(성, 이름): # 2. 함수의 이름을 정하며 필요한 매개변수를 생각한다.
    # 1. 어떤 내용의 함수를 만들지 글로 먼저 정리한다.
    # 전달받은 두 개의 매개변수의 값을 하나로 합쳐서
    # 반환하는 함수

    # 반환할 값을 담을 함수 안의 변수
    result = None
    result = 성 + 이름
    return result # 3. 반환값이 있는지를 생각하고 변수 result를 만든다.

# ========
# if-name-main 코드 블록
if __name__ == "__main__":
    # 함수의 사용
    # 함수명(전달할값)
    # 함수명을 작성한 뒤 소괄호()를 써야 함수를 사용, 호출(call)하는 것이다.
    # 매개변수가 있는 경우에는 반드시 값을 전달해주어야 한다.
    print_name() # 매개변수가 없다면 값을 전달하면 안된다.
    # 매개변수x, 반환값x, 단순히 출력만 하는 기능을 수행한다.
    print_text("전달된 인자, 인수 하나")

    print("===변수o, 반환o===")
    result = make_full_name("이", "준상")
    # 함수에서 return을 통해 반환되는 값이 있는 경우,
    # 함수를 사용한 위치에 반환된 값이 치환되어 나오게 된다.
    # 즉, return이 있는 함수는 하나의 값으로 취급할 줄 알아야 한다.

    print(result)
    pass
# ========
```
