```python
# part3_method_default.py
# 함수를 호출하며 인자, 인수를 넘기지 않았을 때
# 기본값을 설정하는 방법
# def 함수명(매개=기본값):
# 매개변수명을 작성하며 그 뒤에 등호로 값을 전달하면
# 외부에서 값을 전달받지 않았을 때, 함수에서 미리 정해놓은
# 기본값이 대신 할당된다.

# 이름을 전달받을 때 나이를 같이 전달받을 건데,
# 나이를 입력하지 않은 경우, -1을 전달하고
# 이름과 나이를 출력하도록 함수 작성
def print_info(이름, 나이=-1):
    # '이름' 변수에는 반드시 값을 전달해주어야 하지만
    # '나이' 변수에는 값을 전달해도 되고, 전달하지 않은 경우 기본값이 저장된다.
    print("이름: ", 이름, "나이: ", 나이)

if __name__ == "__main__":
    print_info("이준상", 32) # 이름과 나이를 모두 전달하는 경우
    # 전달된 값을 사용한다.
    print_info("이준상2") # 나이를 전달하지 않은 경우
    # 기본값인 -1이 '나이' 변수에 전달된다.
```
