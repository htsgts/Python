```python
# part7_list_comprehension.py
# 리스트 컴프리헨션
# 여러 요소에 대해서 동일한 동작을 수행하여 새로운 리스트를 만드는
# 함수형 프로그래밍을 적극반영한 파이썬만의 문법이다.
# 이 리스트 컴프리헨션은 리스트를 쉽게 만들 수 있도록 도와준다.
# 기본문법
# 일반 리스트를 만드는 것처럼 대괄호 안에 작성한다.
# [e for e in iter]
# 요소를 담을 임시변수 e에 대해서 iter라는 이터러블 객체를 활용하는 내용
# 첫 부분의 e 부분에 어떤 동작을 작성하면 해당 결과값이
# 새로운 리스트의 요소로 들어간다.

def list_comp(input_list: list[int], condition:bool=True):
    # list[int]는 요소가 모두 정수인 리스트를 가리킨다.
    made_list = [e for e in input_list] # 어떠한 조작도 가하지 않은 리스트 생성
    made_list = [(e + 10) for e in input_list]
    print("made_list: ", made_list)
    print("input_list:", input_list)
    # 조건식을 추가하여 해당 조건식을 만족하는 요소만 새로운 리스트의 요소로 편입시킨다.
    # [e for e in iter if 조건식]
    made_list = [e * 3 for e in input_list if (e % 2 == 0)]
    print("조건식 적용:", made_list)

if __name__ == "__main__":
    list_comp([1, 2, 3, 4])
```
