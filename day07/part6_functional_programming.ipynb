```python
# part6_functional_programming.py
# 함수형 프로그래밍이란, 외부상태를 변경하지 않는 순수함수를 조합하여
# 기능을 구현하는 패러다임을 함수형 프로그래밍이라고 한다.
# 주로 많은 데이터를 다루는 이터러블 객체, 시퀀스 등에 대해서
# 모든 요소에 동일한 동작을 수행하게끔 할 때 사용된다.

# 파이썬에서 제공하는 함수형 프로그래밍 도구(함수)는
# 대표적으로 세 가지가 있다.
# 1. map()
# 각 요소마다 동일한 동작을 수행하게 만드는 함수
def test_map():
    # 모든 요소가 숫자로 이루어진 문자열인 경우
    my_list = ["1", "3", "13"]
    print(my_list) # 문자열 상태의 리스트
    # 모든 요소를 숫자로 바꾸고 싶다.
    # map()을 이용해서 int() 형변환 함수를 적용해보자.
    # map(함수명, 시퀀스)
    result = map(int, my_list)
    print(list(result)) # 숫자로 바뀐 리스트

# 2. filter()
# 각 요소를 검사하여 조건식을 통과하는 요소만 걸러내어 새로운 리스트를 만드는 함수
# 반환값이 논리자료형인 함수와 시퀀스를 전달하여 사용할 수 있다.
# 각 요소는 해당 함수에 전달되어 결과값이 나오면
# True인 경우에만 새로운 리스트의 요소로 전달된다.
# filter(함수, 시퀀스)
def test_filter():
    my_list = [1, 2, 3, 4, 5]
    # 람다식(lambda)
    # 함수를 짧게 만드는 방법
    # lambda 매개변수: 반환값
    # 짝수인지 여부를 반환하는 람다 함수
    is_even = lambda x: x % 2 == 0
    filtered_list = filter(is_even, my_list)
    print(list(filtered_list))
    pass
# def is_even(x):
#     return x % 2 == 0
# 3. reduce()
# 요소를 소모하여 하나의 연산결과를 만들어내는 함수
# 예시: 각 요소를 더해서 최종 결과값 하나를 만들어내는 기능
# 기본 내장 함수가 아니라서 import를 해주어야 한다.
from functools import reduce

def sum_element(my_list=None):
    result = None
    my_list = my_list or list(range(1, 10+1)) # 1 ~ 10
    # 1부터 10까지의 합은 55다
    # reduce는 각 요소를 순차적으로 두 개의 매개변수에 전달하여
    # 그 결과값을 다음 반복 때 첫번째 매개변수로 저장한다.
    # 해당 반복 때는 다음 요소가 두번째 매개변수로 저장되어 연산된다.
    # 함수는 두 개의 매개변수를 받아야 하며,
    # 두번째 인자로는 이터러블 객체를 전달하여
    # 요소들을 사용할 수 있도록 해야 한다.
    result = reduce(lambda x, y: x+y, my_list)
    # reduce의 결과값은 하나만 반환된다.
    return result

if __name__ == "__main__":
    # test_map()
    # test_filter()
    result = sum_element([1, 3, 5, 7]) # 16
    print(result)
    pass
```
