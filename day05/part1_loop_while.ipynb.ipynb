## 기타 제어문
- if문이나 while문은 콜론(:)을 통해서 그 영역을 지정한다.
- 이러한 기본흐름을 바꾸는 제어문뿐만 아니라, 이러한 제어문 안에서 흐름을 바꾸는 기타 제어문 두 가지를 배웠다.
- `break`는 for나 while 등의 반복문 흐름 중간에 사용하여 반복은 중단하는 역할을 한다.
- `continue`는 다음 반복으로 이동하여 continue 아래쪽의 코드를 건너뛰는 역할을 한다.
- **공통점**: 두 기타 제어문 모두 해당 코드를 만나는 순간 다른 위치로 흐름을 보내버리기 때문에, 해당 코드 아래쪽의 반복문 안쪽에 위치한 코드는 실행되지 않는다.

````python
# 실습문제1. 외부에서 count 변수를 선언하여 0을 대입
count = 0

# count변수의 값이 100 미만인 동안 print()로 count 출력
while count < 100:
    # count 변수의 값을 1 증가 시키는 코드
    count += 1
    # 출력하는 코드 위에
    # '만약 count를 2로 나눴을 때 나머지가 0'이라는 조건을 만족하면 continue 실행
    if count % 2 == 0:
        continue
    # 출력문 위에 if문을 이용하여 count의 값이 80 초과이면 break 실행
    if count > 80: break
    print(count)


# while문 밖, 아래에 count를 한번 출력
print("while문 밖 출력: ", count) # 81
# 예상
# 1, 3, 5,...,77, 79
# 밖 81
````

````python
# 실습문제2.
# input_text를 선언하여 빈문자열"" 대입
# input_text = ""

# while문으로 True 조건식을 주어 무한반복
while True:
    # 내용을 입력받는다
    input_text = input("숫자 혹은 exit 입력")

    # 입력한 내용이 'exit'라면 반복문 종료(break)
    if input_text == 'exit': break

    # 반복문 안에서 입력한 값이 짝수/홀수인지를 검사하여 만약 짝수면
    if int(input_text) % 2 == 0: # 짝수면
        print("짝수")
    # 짝수가 아니라면 홀수 출력
    else: print("홀수")
print(input_text) # 탈출하려면 exit를 입력해야 하므로 항상 exit를 출력
````

## 반복문 for
- 리스트나 튜플, 세트, 딕셔너리 등 반복가능한 객체(iterable)의 요소를 하나씩 가지고 와서 임시 변수에 담아 모든 요소에 대해서 같은 동작을 반복하는 반복문이다.
- 혹은 숫자의 일정 범위를 지정해서 해당 범위의 숫자를 반복적으로 사용하는 반복문을 가리킨다.
- 요소의 개수만큼 반복이 되기 때문에 while문과 다르게 반복횟수를 알 수 있을 때 사용한다.

````python
# for를 이용하는 첫번째 방법
# 리스트에 담긴 요소를 임시 변수에 담아 사용한다.
# for 임시변수 in 반복 객체:
    # 임시변수를 활용한 반복문
for element in ["홍길동", "고길동", "둘리"]:
    print(element) # 리스트에 담긴 요소를 하나씩 들고 와서 사용된다.
    # 다음 반복 전까지 해당 요소는 임시변수에 담겨 활용될 수 있다.
scores = [80, 75, 90, 85]
# names = ["철수", "영희", "길동", "희동"]
for score in scores:
    print("점수:", score)

# 각 요소의 인덱스(순서값)을 알 필요는 없지만
# 앞 뒤 요소의 참조를 할 수가 없다.
# -> for 반복문 안에서 앞뒤 인덱스를 알 방법이 없다.
````

````python
# range() 함수를 이용하여 등차수열의 값을 임시 변수에 담아 사용한다.
# 해당 리스트의 길이를 len() 함수를 이용하여 가져와서
# 등차수열의 끝값으로 설정하고 등차를 1을 주어 0부터 요소에 접근한다.
for i in range(0,len(scores)): # 등차는 기본 1
    # range()는 시작값과 끝값을 설정하면 시작값부터 끝값-1까지의 등차수열을 생성한다.
    print(i) # 0, 1, 2, 3
    element = scores[i] # i는 0부터 3까지 순차적으로 바뀐다.
    # 변수 element에는 scores의 요소 값이 담긴다.
    print(element) # 요소 값을 출력
````

````python
scores = [80, 75, 90, 85]
names = ["철수", "영희", "길동", "희동"]
# range를 이용해서 순서값을 생성해 요소에 접근하면
# 여러 리스트의 요소를 같이 가져올 수 있다.
for i in range(len(scores)): # 값을 하나만 전달하면 끝값만 설정된다.
    # 시작값은 0이 기본
    print(f"{names[i]}: {scores[i]}")
````

````python
# for문과 range 함수를 활용해서 구구단을 출력해보자.
# 구구단 중 2단을 range 함수를 이용하여 출력
for 곱 in range(1, 9+1):
    print(f"2 x {곱} = {2 * 곱}") # 2단 출력 끝
````

## 중첩 반복문
- 반복문 안에 반복문을 작성하면 그것을 중첩 반복문이라고 한다.
- 중첩 반복문은 외부 반복문 안에 또다른 내부 반복문이 동작함으로써, 내부 반복문은 외부 반복 횟수 x 내부 반복 횟수 만큼 반복되는 형태를 가진다.
- 또한 외부 반복이 되는 순간, 내부 반복문은 초기화되어 처음부터 반복을 시작한다.

- 시계를 생각하면 쉬워진다.
- 시, 분, 초는 각각 12번, 60번, 60번의 반복문 안에 들어가는 형태를 가진다.
```python
for 시 in range(12):
    print(f"==={시}시===")
    for 분 in range(60):
        print(f"{시}시 {분}분")
```

````python
# 중첩 반복문을 구구단에 적용해보자.
# 2단부터 9단까지 각각 1곱부터 9곱까지 출력하는 중첩 반복문 구현하기
# 출력 예시:
# ===2단===
# 2 x 1 = 2
# 2 x 2 = 4
# ...
# 2 x 9 = 18
# ===3단===
# 3 x 1 = 3
# ...

# range(2, 9+1) 단
for dan in range(2, 9+1): # 2단부터 9단까지
    print(f"==={dan}단===")
    # for dan in range(2, )
    # range(1, 9+1) 곱
    for gop in range(1, 9+1): # 1부터 9곱까지
        print(f"{dan} x {gop} = {dan * gop}")
````

````python
# 실습4 n단부터 m단까지 출력(n, m의 순서, 크기와 무관하게 동작)
# n단부터 m단까지 오름차순으로 진행
# 1. 둘 중 큰 수와 작은 수를 구분하여 새로운 변수에 담아서 동작실행
n = int(input("정수1 입력: "))
m = int(input("정수2 입력: "))
big = n if n > m else m
small = n if n < m else m
for dan in range(small, big+1):
    print(f"==={dan}단===")
    for gop in range(1, 9+1):
        print(f"{dan} x {gop} = {dan * gop}")
````

````python
# 2. 기존 변수를 활용하여 range에 적절하게 담아서 활용
n = int(input("정수1 입력: "))
m = int(input("정수2 입력: "))
dans = None # range() 함수 결과를 담을 반복 가능한 객체
# 조건에 따라 range 함수를 다르게 설정
if n > m:
    dans = range(m, n+1)
else:
    dans = range(n, m+1)

# 한줄 쓰기
# dans = range(m, n+1) if n > m else range(n, m+1)

# 위에서 설정된 dans를 활용
for dan in dans:
    print(f"==={dan}단===")
    for gop in range(1, 9+1):
        print(f"{dan} x {gop} = {dan * gop}")
````

````python
# 실습4 심화 n단부터 m단까지 5, 2 -> 5단, 4단, 3단, 2단 순서로 출력
# range() 함수의 결과를 list에 담아서 활용할 수 있다.
# 즉, 리스트 뒤집기 reverse 함수를 이용하면 작은 값부터 큰값의 순서를
# 뒤집어서 for문을 사용할 수 있다.
n = int(input("정수1 입력: "))
m = int(input("정수2 입력: "))
big = n if n > m else m
small = n if n < m else m

# range함수를 작성
# 이대로는 오름차순으로 생성된다.
dans = range(small, big+1)

# reverse 기능 사용을 위해 list로 변환
dans = list(dans)

# 뒤집기 위해서 reverse를 사용한다.
dans.reverse()

# 내림차순 구구단 완성
for dan in dans:
    print(f"==={dan}단===")
    for gop in range(1, 9+1):
        print(f"{dan} x {gop} = {dan * gop}")
````
