## 코랩 환경에서 OS 명령어 사용법
- 코랩에서 파이썬 코드 뿐만 아니라, 파이썬이 실행되는 환경인 리눅스 명령어를 사용할 수 있다.
- 리눅스 명령어를 사용하려면 코드 셀에 `!`를 작성한 뒤 그 뒤에 리눅스 명령어를 작성하면 된다.
- 코랩 환경의 파이썬 버전을 확인해보자.
```
!python --version
!python -v
```

````python
# part1_start.py # 항상 파일을 작성할 때에는
# 가장 위에 파일명을 작성해준다.

# .py 파일은 실행하면 파일 통째로 실행이 된다.
# 그래서 코드를 작성하는 영역에 대한 기본 형태가 있는데
# 상단은 필요한 정보를 가져오는 import 영역
import os
import sys

# 중단은 실제로 코드를 작성하는 공간


# 하단은 코드 실행을 시작하는 진입점
# if __name__ == "__main__":
# 위 if문을 작성한 아래쪽의 코드는
# 현재 이 파일을 직접 실행했을 때만 동작, 실행되는 코드다.
if __name__ == "__main__":
    print("Hello world")
````
