## 리스트 요소의 제거
- `.append()`, `.insert()`, `.extend()` 등이 요소를 추가하는 메서드들이었다면
- 요소를 제거하는 메서드는 `.remove()`이다.
```python
nums = list(range(10)) # 0 ~ 9
# 리스트명.remove(인덱스)
# 특정 인덱스의 요소를 제거하고 뒤의 요소를 땡겨 자리시킨다.
nums.remove(3)
# 3이 제거된 nums
print(nums) # [0, 1, 2, 4, 5, 6, 7, 8, 9]
```
- 인덱스를 전달해서 특정 위치의 요소를 제거할 수도 있지만, 가장 마지막 요소(-1)를 제거하고 싶다면, `.pop()`를 사용할 수도 있다.
- 이때 `.pop()`를 사용하면 제거된 요소의 값을 반환한다.
- `.pop(인덱스)`처럼 인덱스 값을 전달하면 해당 위치의 요소를 제거하고 그 값을 반환할 수 있다.
    - 만약 그 위치에 값이 없거나 리스트의 길이를 넘어가는 인덱스를 전달하면 `None`을 반환한다.

````python
# append, insert, extend, remove, pop
# 위의 요소 추가/삭제 메서드는 원본 리스트의 요소를 직접 조작한다는 점에서
# 함수를 사용할 때 조심해서 사용해야 할 함수들이다.
# 직접 조작하는 메서드들은 반환값이 일반적으로 없다.
# 하지만 조작한 결과 리스트를 새로 생성하는 메서드들도 존재한다.

# 원본 리스트를 복사해서 새로운 리스트를 만드는 함수
# .copy()
nums = [1, 2, 3, 4]
# nums를 복사한 새로운 리스트 copied
copied = nums.copy()

# 복사된 copied의 요소를 조작해보자.
copied.pop()
print("copied: ", copied)
print("origin: ", nums)
````

## 리스트의 기타 유용한 함수들
- clear(): 모든 요소를 제거. 변수 이름만 남게 된다.
- sort(): 해당 리스트의 요소를 정렬(기본 오름차순)
- sort(reverse=True): 내림차순 정렬

    - 예) [1, 3, 2, 4] -> [4, 2, 3, 1]

    ```python
    nums = [1, 2, 2, 1, 3, 4, 2]

    ```
- index(): 전달된 값이 처음으로 등장하는 인덱스를 반환. 찾지 못하면 -1 반환

````python
# index() 실습
nums = [1, 2, 2, 3, 1, 4, 1]
# index()를 이용해서 4가 처음으로 등장하는 인덱스를 반환
catched = nums.index(4) # 5
# 찾은 인덱스 출력
print(catched)

# 찾은 인덱스는 해당 값을 pop해서 제거하는 데 쓰일 수 있다.(or 출력)
# nums.pop(catched)
print(nums)

# 찾으려는 요소가 존재하지 않으면 에러가 발생한다.
# 찾으려는 요소가 리스트에 존재하는지 여부를 반환하는 연산자
# in
has_num = (4 in nums) # 4라는 값이 nums의 요소로 존재하니?
print(has_num)

# 만약에 has_num이 True라면 4라는 값이 존재하는 것이므로
# 4의 인덱스를 추출할 수 있다.
if has_num: # 만약 해당 값을 포함한다면
    catched = nums.index(4)
    print(catched)
````

## 리스트 요소의 정렬과 뒤집기
- sort(): 숫자는 오름차순, 문자는 사전순으로 정렬(원본 리스트를 조작 -> 반환x)
- reverse(): 정렬여부와 무관하게 원본 리스트의 요소 순서를 반전시킨다.

````python
# 원본 리스트 생성
nums = [10, 12, 14, 4, 3, 9]
# nums의 복사본을 생성하여 정렬
copied = nums.copy()

# copied 정렬
copied.sort()

# 정렬된 copied 출력
print(copied)

# 내림차순으로 정렬
copied.sort(reverse=1) # True 혹은 1을 주어서 역순 활성화

# 내림차순 정렬된 copied 출력
print(copied)
````

````python
# 요소의 개수 확인하기
copied = nums.copy()
copied.extend([14, 4])
print(copied)
# .count(세고자하는 값)
counted = copied.count(14) # 2
print(counted)

if copied.count(14) != 0: # 개수가 0이 아니라면 있는 거니까
    copied.remove(14) # 실행됨
if copied.count(14) != 0: # 개수가 0이 아니라면 있는 거니까
    copied.remove(14) # 실행됨
if copied.count(14) != 0: # 개수가 0이 아니라면 있는 거니까
    copied.remove(14) # 14가 없어서 실행 안 됨
print(copied)
````

## 리스트의 내장 함수

```python
nums = [1, 2, 3, 4, 5]
```
- len(): 리스트의 길이를 반환(요소의 개수를 반환)
```python
print(len(nums)) # 요소의 개수 5
```

```python
nums = [3, 5, 1, 2, 4]
# 원본을 복사한 복사본으로 정렬한 결과를 반환한다.
print(sorted(nums)) # [1, 2, 3, 4, 5]
# 원본을 직접 수정하지 않는다.
print(nums) # [3, 5, 1, 2, 4]
```

## while 반복문
```python
while 조건식:
    실행할코드
```
- 기본적인 진입 조건과 구조는 if문과 동일하다.
- 다만, if문은 실행한 뒤 아래로 쭉 흘러갔다면, while문은 다시 조건을 검사하러 올라간다는 부분이 다른 점이다.
- 코드 흐름에서 다시 위로 올라가 코드를 실행하며 내려온다는 것은, 같은 코드를 반복한다는 의미가 된다.
- 하지만 코드를 반복한다고 해서 항상 같은 내용으로만 진행되는 것은 아니다.
    - 왜? 변수의 존재 때문
    - 변수는 반복문이 진행되는 상황에서도 자신이 저장한 값은 계속 바뀔 수 있다.
- 이러한 변수의 변화를 기준으로 반복문의 종료 여부가 결정된다.


````python
# while문 기본 문법
num = int(input("숫자를 입력해주세요: "))

# num에 담긴 값이 양수인 동안 반복실행
# num에 담긴 값이 0 또는 음수가 되는 순간 반복 종료
while num > 0:
    # num에 담긴 값을 출력
    print(num)
    # num 변수에 담을 값을 변경
    num = int(input("숫자를 입력해주세요: "))
    # 반복문이 끝나면 위의 조건식을 검사하러 다시 올라간다.
````

````python
# while문 기본 문법
num = int(input("숫자를 입력해주세요: "))

# num에 담긴 값이 양수인 동안 반복실행
# num에 담긴 값이 0 또는 음수가 되는 순간 반복 종료
while num > 0:
    # num에 담긴 값을 출력
    print(num)
    # num 변수에 담을 값을 변경
    num = int(input("숫자를 입력해주세요: "))
    # 반복문이 끝나면 위의 조건식을 검사하러 다시 올라간다.

# while문은 조건식이 만족하는 동안 **무한히** 반복실행한다.
# 즉, 조건식을 만족하지 않는 경우를 만들지 않으면
# 무한 반복에 빠져 컴퓨터가 망가질 수 있다. *자동으로 중단된다.

# 만족하지 않는 경우, 즉, 끝값을 설정하는 방법뿐만 아니라
# 특정 입력을 통해서 임의로 반복을 종료하는 방법도 존재한다.
````

````python
# 기타 제어문
# while문의 조건식을 항상 참(True)으로 설정한 뒤
# 내부에서 조건문if를 이용하여 특정 조건이 되면
# 반복문을 탈출(break)하는 방식을 사용할 수 있다.
# 이때 반복문을 종료하는 예약어를 break라고 하며
# 만나면 다음 반복으로 보내버리는 예약어를 continue가 있다.

# 무한 반복
while 1: # while True:
    value = input("아무거나 입력해주세요(종료: exit 입력): ")
    print(value)
    if value == "exit": # 입력된 문자열이 "exit"이면
        # 반복을 종료(탈출)
        break
        # break를 만난 이후의 코드는 접근할 수가 없다.
    else: # 입력된 값이 "exit"가 아니면
        # 다음 반복으로 이동
        continue
        # 이후의 코드 실행 불가
    print("실행될 수 없는 코드") # 닿을 수 없는 코드(unreachable)
````
