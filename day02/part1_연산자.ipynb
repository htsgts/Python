- 연산 우선순위를 모르겠으면 소괄호(최우선 연산자)를 이용해서 묶음을 명시적으로 표시해 가독성을 끌어올릴 수 있다.

````python
# 산술 연산자
# 사칙연산(덧셈, 뺄셈, 곱셈, 나눗셈)
# +, -, *, /
# 사칙연산의 결과값은 정수 혹은 실수로 나온다.
num1 = 3.5
num2 = 4
print(num1 + num2) # 7.5
print(num1 - num2) # -0.5
print(num1 * num2) # 14? 14.0? -> 14.0
print(num1 / num2) # 0.875

# 피연산자 중 하나라도 실수라면 그 결과값은 실수로 도출된다.
# 자동형변환: 연산을 할 때 작은 크기의 자료형이 큰 타입의 자료형으로
# 별다른 조치없이 자동으로 형변환되는 것을 자동형변환이라고 한다.
````

````python
# 비교(관계) 연산자
# 두 값의 관계를 비교하여 참과 거짓으로 반환하는 연산자를 가리킨다.
# 같니? 다르니?
# ==    !=
# 같거나 다른지 '여부'(T/F)를 반환한다.
print("3 == 3", 3 == 3) # 3이랑 3이 같니? -> 예 -> True
print("3 != 3", 3 != 3) # 3이랑 3이 다르니? -> 아니오 -> False

num1 = 3
num2 = 10
# 두 숫자값의 크기를 비교하는 연산자
# 초과 미만
# >,   <
# A > B -> A가 B보다 크니? -> 예/아니오 -> T/F
print("3 > 10", num1 > num2) # num1이 num2 초과이니? -> 아니오 -> False
print("3 < 10", num1 < num2) # num1이 num2 미만이니? -> 예 -> True

# 이상, 이하
# 초과 미만에서 해당 경계선에 위치한 숫자값을 포함하는 연산자
# >=, <=
print("3 >= 10", 3 >= 10) # 3이 10보다 크거나 같니?
print("3 <= 10", 3 <= 10) # 3이 10보다 작거나 같니?
print("5 <= 5", 5 <= 5)   # 5가 5보다 작거나 같니?
````

````python
# 논리 연산자
# 논리값과 논리값을 연산하여 논리값을 반환하는 연산자
# 논리곱: and
# 두 조건식(T/F)이 모두 만족(True)하면 그 결과 True를 반환하는 연산자
# * 하나라도 만족하지 않으면(False) 그 결과로 False를 반환한다
# 논리곱이라고 부르는 이유는 조건식의 값을 1 또는 0으로 치환하고
# and 연산자를 곱셈으로 바꿨을 때 그 결과가 같기 때문이다.
# (T and T) == T
# (T and F) == F
# (F and F) == F
print(True and True) # 1 * 1 == 1(True)
print(True and False)# 1 * 0 == 0(False)
print(False and False)#0 * 0 == 0(False)

print("=" * 10) # 문자열 = 10번 반복

# 논리합: or
````

````python
# 논리 연산자
# 논리값과 논리값을 연산하여 논리값을 반환하는 연산자
# 논리곱: and
# 두 조건식(T/F)이 모두 만족(True)하면 그 결과 True를 반환하는 연산자
# * 하나라도 만족하지 않으면(False) 그 결과로 False를 반환한다
# 논리곱이라고 부르는 이유는 조건식의 값을 1 또는 0으로 치환하고
# and 연산자를 곱셈으로 바꿨을 때 그 결과가 같기 때문이다.
# (T and T) == T
# (T and F) == F
# (F and F) == F
print(True and True) # 1 * 1 == 1(True)
print(True and False)# 1 * 0 == 0(False)
print(False and False)#0 * 0 == 0(False)

print("=" * 10) # 문자열 = 10번 반복

# 논리합: or
# 논리곱과 마찬가지로 논리값(0/1)로 치환했을 때, 두 논리값의 합을 연산하는 연산자를 가리킨다
# 합집합과 개념을 같이한다.
# 둘 중 하나라도 참이면 그 결과 True를 반환한다.
# *가장 먼저 True로 판명되는 값을 반환한다.
# (True or True) == True
# (True or False) == True
# (False or False) == False
print(True or True) # 1 + 1 == 1(True)
print(True or False)# 1 + 0 == 1(True)
print(False or False)#0 + 0 == 0(False)

# 가장 먼저 True로 판명되는 값을 반환하는 예시
print("0" or "홍길동") # "0"은 내용이 있는 문자열이므로 True로 판단한다.
# 그 결과 "0"이 출력된다.
print(None or 30) # None은 False로 취급되므로 두번째 조건인 30을 반환한다.
````

````python
# 나누기의 소수점 표현 범위
# 15번째 자리까지 정확히 표현된다.
# 16번째 소수점자리까지 표현되지만 부정확하다.(근삿값)
print(10 / 3) # 3.33333
# 소숫점 아래 n번째 자리까지 표현하고 싶을 때 사용하는 서식문자
# 중괄호{} 안에 표현하려는 값을 작성하고 콜론:과 . 그리고 표현할 자릿수를 적고
# f를 적어서 float임을 표현한다.
print(f"{(10 / 3):.5f}")
# 소숫점 자리 이동하기
print(f"{3.141592123412341234 * (10 ** (-3))}")

# 정수를 문자열로 표현할 때 일정한 자릿수만큼의 공간을 미리 차지하고 싶을 때
print(f"{1234:5d}") # 5자리 만큼의 자리를 미리 차지하고 숫자값을 오른쪽정렬(기본값)
# 자릿수 앞에 0을 작성하면 빈 공간을 0으로 채워넣는다.
print(f"{1234:05d}")
# 왼쪽 정렬은 콜론: 뒤에 <, 오른쪽 정렬(기본값)은 >
print(f"{1234:<05d}") # 12340
````

## 삼항 연산자
### 조건식(condition)
- 조건식이란, 연산자를 이용한 연산식(표현식)의 결과가 True 혹은 False로 반환되는 식을 가리킨다.
- '조건에 따라'라는 말은 조건식의 결과값이 True냐 False냐에 따라 분기가 생기는 것을 가리킨다.
### 삼항연산자
- (참일때 값) if 조건식 else (거짓일 때 값)
```python
num1 = 3
num2 = 5
result = "num1이 더 큽니다."if num1 > num2 else "num2가 더 큽니다"
print(result)
```
## 대입(할당) 연산자
- 단순할당 연산자
    - `=` 우항의 값을 좌항의 변수에 저장하는 연산자
    - 모든 연산자 중 가장 마지막에 연산되며 연산 방향이 오른쪽에서 왼쪽이다.
    - 기존에 존재하던 변수에 값을 할당하는 경우, 기존값은 사라지고 새로운 값이 저장된다.(덮어쓰기)
- 복합할당 연산자
    - 배경: 기존의 단순할당 연산자는 원래의 값과 무관하게 새로운 값을 덮어쓰기 하는 기능만 있었다.
        - 때문에 기존 값을 활용하려면 아래와 같이 작성해야 한다.
        ```python
        age = 30
        # 1살 먹음
        age = age + 1
        # 위 코드에서 변수가 두 번 반복되어 호출되기 때문에 코드가 길어진다.
        # 변수 이름을 두 번 쓰지 않으려고 복합할당 연산자를 만든다.
        ```
    - 기존의 변수에 저장된 값을 산술연산자의 **좌항으로 취급**하고 복합할당연산자의 우항을 산술연산자의 **우항으로 취급**하여 **산술연산 후 할당까지** 한꺼번에 하는 연산자이다.


````python
# 변수를 선언, 초기화 후 해당 값을 3 증가시키는 코드
age = 30
# 기존 단순할당방식
age = age + 3 # 33

# 복합할당방식
age = 30
age += 3 # 변수를 한번만 사용해도 된다.
# 33

amount = 10 # 기존 개수가 10개
# 기존 개수를 3배로 늘려 해당 변수에 연산된 값을 다시 저장
amount *= 3 # 3배로 늘린다
print(amount) # 30

# 그외 복합할당연산자
# -= /= //= %= **= 등이 존재한다.

# 함수의 소괄호() 안에서 할당을 한 뒤 그 변수의 값을 사용하려면
# 콜론: 등호=를 사용해야 한다.
print(name:="홍길동") # 할당과 사용을 동시에 한다.
print(name) # 사용(값의 호출)

# 대입 연산자를 연쇄적으로 사용하면 같은 값을 여러 변수에 넣을 수 있다.
num3 = num2 = num1 = 3
print(num1, num2, num3)

# 반점,을 이용한 여러 값 동시 할당
# 할당연산자 좌우로 같은 개수의 값이나 변수를 반점으로 구분하여 나열하면
# 순서에 맞게 우항의 값이 좌항의 변수에 할당된다.
num1 = 10
num2 = 20
num1, num2 = 21, 31
print(num1, num2) # 21, 31

# 만약 두 변수의 값을 서로 바꾸고 싶으면 동시 할당 방법을 이용해서 swap이 가능하다.
num1 = 10
num2 = 20
# 두 변수에 담긴 값을 서로 바꾼다.
num1, num2 = num2, num1
print(num1, num2) # 20 10

# 위 방식을 사용하지 않고 서로 값을 바꾸려면 임시 변수가 하나 필요하다
num1 = 10
num2 = 20
temp = 0 # 임시 변수

temp = num1 # temp에 num1에 담겨 있던 값을 복사해 넣는다
num1 = num2 # num2에 있던 값 20을 num1에 복사해 저장한다.
num2 = temp # temp에 담겨 있던 num1의 값 10을 num2에 복사해 저장한다.
# ** 더 이상 사용하지 않는 변수를 제거
# del 변수명
del temp
print(num1, num2) # 20 10
# print(temp) # 에러 발생(변수 선언하지 않음)
````

## 연산자 심화
### 동등 연산자의 의미
- 동등연산자(==)는 두 값이 같은지 여부를 반환하는 연산자라고 했다.
- 그렇다면 문자열 "3"과 숫자 3은 서로 같다고 판정할까?
```python
print(3 == "3")
```

````python
print(3 == "3") # 같지 않다
# 동등 연산자는 해당 값의 내용뿐만 아니라 타입도 검사한다.
# 정수와 실수의 동등 여부는 어떻게 판단될까?
print(3 == 3.0) # 두 값 중 하나라도 실수라면 둘 모두 실수로 변환되어
# 연산되므로 내부적으로 3.0 == 3.0으로 연산된다. -> True
````

## 특수 연산자의 종류
- `is`: 동등연산자와 마찬가지로 동등여부를 반환하는 연산자이다.
- 값뿐만 아니라 같은 저장공간(객체)인지 여부까지도 검사를 한다.
```python
num1 = 10
num2 = 10
print(num1 is num2) # False
```

````python
num1 = 10
num2 = 10
print(num1 is num2) # 단순 변수끼리는 값을 비교한다.
# is는 단순 값(Literal: 기본자료형으로 이루어진 값)을 비교하는 것이 아니라
# 같은 객체인지를 검사하는 연산자이다.

# 그런데 리터럴값은 미리 준비된 값을 변수가 참조하는 형태이기 때문에
# is를 사용했을 때 같다는 결과가 나온다.
````
