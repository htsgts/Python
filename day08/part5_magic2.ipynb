```python
# part5_magic2.py
# 매직 메서드 중, 연산자를 재정의하는 내용
# - __add__(self, other) : 현재 객체와 다른 객체 간에 덧셈 연산+을 했을 때의 동작을 정의
# - __sub__(self, other) : 뺄셈
# - __mul__(self, other) : 곱셈
# - __truediv__(self, other) : 나눗셈
# 일반적으로 모든 클래스는 연산자 매직 메서드가 정의되어 있지 않다.
# 때문에 연산자 매직 메서드를 사용하고자 한다면
# 해당 메서드를 재정의 해줘야 한다.

# 예제: 좌표 클래스로 연산자 메서드 연습
class Coordinate:
    # x와 y 좌표를 저장하는 클래스
    def __init__(self, x: int, y: int):
        self.point = (x, y)

    # 좌표 클래스 객체 간의 산술연산에 대해서
    # 매직 메서드로 정의를 해보자.
    def __add__(self, other):
        x = self.point[0] + other.point[0]
        y = self.point[1] + other.point[1]
        result = Coordinate(x, y)
        return result

    def __str__(self) -> str:
        return f"x: {self.point[0]}, y: {self.point[1]}"
    def __repr__(self):
        return "대표 문자열"
if __name__ == "__main__":
    point1 = Coordinate(3, 5)
    point2 = Coordinate(7, 5)
    # 두 좌표 객체를 더하면
    # x좌표의 합을 x좌표값으로 하고
    # y좌표의 합을 y좌표값으로 하는 새로운 좌표 객체를 반환
    point_added = point1 + point2
    print(point_added.point) # (10, 10)

    # __str__을 정의하면 해당 객체를 str() 안에 넣었을 때 나오는 문자열을 정할 수 있다.
    print(str(point_added))
    print(repr(point_added))
    print(point_added)
    # __str__로 정의한 내용은 __repr__로 정의한 것보다 우선순위가 높다.
    # 때문에 일반적으로 __repr__를 정의하는 것보다
    # __str__만 하나 재정의하는 것이 일반적이다.


```
