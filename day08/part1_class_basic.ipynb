```python
# part1_class_basic.py
# 클래스(Class)
# 클래스란 양산을 위한 양식을 만드는 틀을 가리킨다.
# 이 클래스라는 양식을 기반으로 수많은 객체를 만들어낼 수 있다.
# 각 객체는 같은 양식, 틀을 공유하기 때문에
# 묶어서 다룰 수가 있게 된다.

# 클래스 만들기
# class 클래스명: (필드)를 채워넣는다.
# class라고 적고 공백을 작성한 뒤 클래스명을 쓰고
# 콜론: 아래에 들여쓰기하여 속성, 함수(메서드) 등을 작성하면
# 해당 내용들은 모두 해당 클래스에 소속된 정보가 된다.

# 사람이라는 클래스(카테고리) 만들기
# 이름, 나이, 성별, 점수들(국,영,수)
class Person:
    # 속성
    # name
    # age
    # gender
    # scores
    # 생성자(initializer). 초기화를 위한 특수 메서드
    # 속성과 메서드는 생성자를 시작으로 만들 수 있다.
    def __init__(self): # 생성자는 반드시 self라는
        # 매개변수를 받아야 한다.
        # 이 self라는 매개변수는 이 클래스를 기반으로 만들어지는 객체들의
        # 속성과 메서드에 접근하기 위한 키(지칭대명사)가 된다.

        # 속성 선언 및 초기화
        # 해당 객체의 정보에 접근하기 위해서는
        # 하위접근연산자인 마침표.를 self 뒤에 작성해야 한다.
        self.name = "이름없음"
        self.age = 0
        return # 생성자는 반환값이 없어야 한다.
        # __init__ 안에서의 return은
        # 생성자를 끝내는 역할만 수행한다.
    # 메서드

if __name__ == "__main__":
    # 클래스의 객체화, 인스턴스화
    # 양식을 기반으로 구체화된 새로운 대상을 만드는 것
    # 객체를 부를 키인 변수명을 작성하고
    # 객체를 해당 변수에 저장한다.
    # 클래스 이름과 함께 소괄호()를 사용하여
    # 생성자 메서드를 호출한다.
    hong = Person() # hong이라는 이름으로 해당 객체에게
    # 접근할 수 있다.
    hong.name = "홍길동" # 객체'의.' name을 재설정
    # .name은 해당 객체에 소속되어 있는 속성이다.
    # 다른 객체와 같은 이름으로 불리지만(name)
    # 실제 저장공간은 서로 다르다.
    kim = Person() # 두번째 객체 생성
    kim.name = "김유신" # kim이라는 객체의 변수 name이기 때문에
    # hong의 name과는 서로 다른 저장공간을 가리키게 된다.
    print(kim.name)
    print(hong.name)
    # 두 객체를 하나의 리스트에 담아서
    # 일괄적으로 관리할 수 있다.
    # why? 같은 양식을 공유하기 때문에 변수명이 서로 같다.
    people = [hong, kim]
    # 반복문으로 각 이름과 나이를 일괄출력
    for person in people:
        print(f"이름: {person.name}\n"+
              f"나이: {person.age}")



    pass
```
