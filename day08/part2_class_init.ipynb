```python
# part2_class_init.py
# 클래스를 객체화할 때 필요한 정보 빠르게 입력하기
# 생성자__init__이 이러한 역할을 담당한다.
class Person():
    def __init__(self, name, age):
        # 생성자는 해당 객체를 인스턴스화할 때
        # 필요한 정보와 세팅을 선제적으로 진행하는
        # 역할을 담당한다.
        # 이 생성자는 클래스명()를 호출할 때 클래스 대신에 호출되어
        # 메서드로서의 역할을 수행한다.
        # 단, self 매개변수는 전달받는 대상이 정해져 있기 때문에
        # 클래스명()를 사용할 때 전달받는 인자의 개수는 self를 제외한 개수이다.
        self.name = name
        self.age = age
        print("이름, 나이 세팅 완료")

if __name__ == "__main__":
    hong = Person("홍길동", 30)
    # 내부에 세팅된 .name 변수의 값을 호출
    print(hong.name) # 홍길동
    print(hong.age) # 30
    # 원시 자료형과 참조 자료형의 차이
    # 원시 자료형이란, 그 자체로 값을 지닌 자료형을 가리킨다.
    # 참조 자료형은 그 자체로 값을 가진 것이 아니라
    # 값을 가진 다른 대상을 가리키고 바라보는 자료형을 말한다.
    # 즉, 참조 자료형은 대상이 사라지면 참조 오류가 발생하는 자료형, 변수인 것이다.
    # -> 클래스에 의해서 만들어진 인스턴스, 객체는
    # 참조 자료형에 속한다.
    의적 = hong # 의적이라는 변수(별명)에 hong의 정보를 전달해준다.
    print(hong.name) # 홍길동
    # 나이 값을 변경
    의적.age = 40
    print("의적의 나이: ", 의적.age) # 40
    print("홍길동의 나이: ", hong.age) # 40

    # ===================
    # 원시 자료형을 품은 변수는 값을 복사한다.
    name = "이준상"
    nickname = name
    name = "이씨"
    print("닉네임: ", nickname)
    print("이름: ", name)
```
