```python
# part4_magic_method.py
# 매직 메서드란, 파이썬의 기본적인 동작과 함수만으로는 구현할 수 없는
# 특수한 동작이나 연산에 대해서 사용자가 마음대로 바꿀 수 있도록 하는 특수한 메서드를 가리킨다.
# 1. 생성, 소멸 메서드
# - __init__ : 새로운 객체를 생성할 때 호출되는 메서드
# - __new__  : 객체 생성시 클래스 정보를 세팅하는 메서드
# - __del__  : 객체를 메모리 해제(del)할 때 동작을 정의하는 메서드
class Person():
    def __init__(self, name):
        print("생성자")
        self.name = name
        pass

    def __new__(cls):
        print("new 메서드, cls: ", cls)
        pass

    # 현재 동작 안함. 버전 문제일 가능성
    # def __del__(self):
    #     print("소멸자 메서드")

if __name__ == "__main__":
    # hong = Person("홍") # __new__가 정의되어 있어서
    # __init__이 호출되지 않는다.

    hong = Person()
    # __new__ 매직 메서드를 사용하면
    # __init__ 생성자를 정의하더라도
    # __new__ 매직 메서드가 먼저 호출되기 때문에
    # 전달하는 인자가 서로 일치하지 않아
    # 오류가 발생할 수 있다.
    del hong
```
