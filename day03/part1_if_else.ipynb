# 삼항 연산자 if-else 복습
- 삼항연산자란, `if 조건식`의 결과에 따라 다른 값이 선택되도록 만드는 사용자 정의 연산자 같은 것이다.
```python
num1 = 30
num2 = 35
# 만약(if) num1이 num2보다 크다면
# 좌항의 num1을 선택하고
# 아니라면(else) 우항의 num2를 선택해라.
# 선택된 값은 big_one 변수에 저장해라.
big_one = num1 if num1 > num2 else num2
```
- 주로 단순 조건식의 결과에 따라 결과값을 다르게 주고 싶을 때 사용하지만, 가독성이 좋지 못한 경우엔 이후에 배울 `if: else:`문으로 풀어서 작성하는 게 좋다.


````python
num1 = 30
num2 = 35
# 만약(if) num1이 num2보다 크다면
# 좌항의 num1을 선택하고
# 아니라면(else) 우항의 num2를 선택해라.
# 선택된 값은 big_one 변수에 저장해라.
big_one = num1 if num1 > num2 else num2
````

````python
# num1의 숫자를 직접 입력하여 35와 비교 후, 더 큰 값을 출력하는 예시
num1 = int(input("숫자를 입력해주세요"))
num2 = 35
# 만약(if) num1이 num2보다 크다면
# 좌항의 num1을 선택하고
# 아니라면(else) 우항의 num2를 선택해라.
# 선택된 값은 big_one 변수에 저장해라.
big_one = num1 if num1 > num2 else num2
print(big_one)
````

````python
# 삼항 연산자 연습
# 1. num1이라는 변수에 30이라는 값을 넣고
num1 = 30
# num2라는 변수에 input() 함수로 입력을 받아 정수로 변환하여 저장
num2 = int(input("정수를 입력해주세요: "))
# 두 값이 같은지(같니?) 여부에 따라 "같습니다" 혹은 "다릅니다"를 변수 text에 저장
condition = (num1 == num2)
text = "같습니다" if condition == True else "다릅니다"
# print(text)를 통해서 결과 출력
print(text)
````

# 제어문(Control Statement)
- 제어문의 종류에는 크게 두 가지가 있다.
1. 조건문
2. 반복문
- 이중에서 조건문의 if-else문을 배워보도록 하자.
## 조건식
- 조건식이란 그 결과값이 논리값으로 나오는 연산식(표현식)을 가리킨다.
- 조건식의 내용이 단순하면 그냥 작성해도 되지만, 여러 조건이 복합적으로 작용하는 복잡한 조건식은 변수에 그 결과값을 담아서 사용하는 것이 가독성에 좋다.
- 이러한 조건식을 구성하는 연산자는 주로 비교연산자와 논리연산자가 있다.
- 파이썬의 경우, 조건식 대신에 단순 값이 들어갈 수 있다.(정수, 실수, 문자열, None 등)
    - 이때 값이 0, "", None의 경우 False로 취급되고, 그외의 값은 모두 True로 취급된다.
    - 이를 이용해서 0인지 여부를 단순 값을 전달하여 확인할 수도 있다.

## 조건문 if
- if문은 조건식을 if 뒤에 작성한 뒤 콜론(:)을 붙여 코드 블록의 시작을 알린다.
```python
name = "홍길동"
if name == "홍길동":
    print(name)
```
- 이러한 코드블록은 해당 들여쓰기가 연속되는 동안 하나의 코드블록으로 묶이며, 조건식이 참이면 해당 코드블록이 실행된다.
- 들여쓰기를 하지 않는 부분이 나오는 순간 if문과는 무관한 외부의 코드로 취급된다.
```python
if name == "홍길동":
    print(name)
print("if문 바깥쪽. 조건식과 무관하게 항상 실행")
```

````python
# if-else
# if문을 두 번 작성하여 조건식을 부정하는 방식은 비효율적이고
# 가독성을 떨어뜨리는 문제를 발생시킨다.
# 앞선 조건식이 참이면 if:의 코드블록을 실행하고
# 조건식이 거짓(False)이면 else: 문의 코드블록을 실행한다.

# 나이를 입력받아서 20세 이상이면 "성인" 출력
age = int(input("나이를 입력해주세요: "))

if age >= 20:
    print("성인")
# 20세 미만이면(성인이 아니면) "미성년자" 출력
else: # 위 조건식이 거짓이면 실행되는 코드블록
    print("미성년자")

# else:문은 여집합이기 때문에 사실상 모든 케이스에 대응할 수 있다.
# 즉, 둘 중 하나의 코드블록은 반드시 실행된다는 것이다.
# 이러한 else:문은 if문을 필요로 하지만 반드시 작성해야 하는 것은 아니다.
# 하지만 if문만 단독으로 사용했을 때 else를 고려했는지를 알 수가 없다.
# else:를 쓸 필요가 없는 경우, else: 코드블록을 작성하고
# '그냥 넘어가세요'라는 의미로 `pass`를 작성한다.
````

````python
# if-else를 사용하여 문제 풀기

# 1. 숫자를 입력받아서 값이 60점 이상이면 "합격"
# 60점 미만이면 "과락"을 출력

# 2. 이름을 입력받아서
name = input("이름을 입력해주세요: ")
# 빈문자열인지를 저장한 변수 is_empty
is_not_empty = bool(name) # 입력받은 문자열을 논리값으로 변환(T/F)

# case1
# 빈문자열이면
if is_not_empty == False: # ""이면
    # "다시 입력해주세요" 출력
    print("다시 입력해주세요")
# 조건과 무관하게 입력받은 값 출력
print(name)

# case2
if name: # 입력받은 값이 있다면
    pass # 그냥 넘어가기
else:
    # 입력받은 값이 없다면
    print("다시 입력해주세요.")
print(name)

# 3. 만약 입력받은 값이
num = int(input("정수를 입력해주세요: "))

# 10 이상이면 "10 이상"
if num >= 10: print("10 이상")
else: # 10이상이 아닌 경우
    # 10 이상이 아닌 경우 중
    # 0 이상이면 "0 이상 10 미만"
    if num >= 0: print("0 이상 10 미만")
    # 10 이상이 아닌 경우 중
    # 0 미만이면 "0 미만" 출력
    else: print("0 미만")
````

````python
# 2. 이름을 입력받아서 빈문자열이면 "다시 입력해주세요" 출력
name = input("이름을 입력해주세요: ")
if name == "":
    print("다시 입력해주세요")

# 조건과 무관하게 입력받은 값 출력
print(name)
````

````python
# elif문
# else문 안에 또다른 조건식을 쓰고자 한다면
# if문과 else문 사이에 elif문을 작성할 수 있다.
# 이는 앞선 조건식이 거짓이라는 조건과 함께
# elif의 조건식이 만족하면 해당 코드블록을 실행한다.
# * if-elif-else문은 가장 먼저 참이 되는 조건식을 가진 코드 블록이 실행된다.

# 입력받은 나이에 따라서 10대, 20대, 30대, 40대 이상을 출력해보자.
# 나이를 입력받아 정수로 변환
age = int(input("나이를 입력해주세요: "))

# 만약에 age의 값이 40 이상이면 "40대 이상" 출력
if age >= 40: print("40대 이상")
elif age >= 30: print("30대")
elif age >= 20: print("20대")
elif age >= 10: print("10대")
elif age >= 0: print("유아")
else: print("나이는 음수가 될 수 없습니다.")
# 잘못 입력된 값에 대한 처리
````

## if-elif 주의사항
- if-elif를 사용할 때에는 여러 조건을 쓰기 때문에 그 순서와 조건식을 주의깊게 고려해야 한다.
- 앞선 조건식이 참인 경우에 뒤의 조건식도 모두 참이라면(포함관계) 뒤의 조건식은 절대 실행될 수가 없다.
    - 순서를 바꿔줘야 한다.
    - 오류가 발생하지 않는다.
1. 조건식의 범위가 겹치지 않도록 조건식을 설정한다.
2. 앞선 조건식의 결과에 뒤의 조건식이 귀속되지 않도록 순서를 조정한다

````python
# 나이를 입력받아 정수로 변환
age = int(input("나이를 입력해주세요: "))

# 만약에 age의 값이 40 이상이면 "40대 이상" 출력
if 40 <= age: print("40대 이상")
elif 20 <= age < 30: print("20대")
elif 30 <= age < 40: print("30대")
elif 10 <= age < 20: print("10대")
elif 0 <= age < 10: print("유아")
else: print("나이는 음수가 될 수 없습니다.")

# if-elif를 사용할 때에는 순서를 명확하게 설정하거나
# 조건식의 범위를 겹치지 않게 설정해야 한다.
````

# 리스트 자료구조
- 리스트 자료구조는 이터러블 자료구조 중 가장 기본이 되는, 가장 자주 사용되는 자료구조이다.
- 순서가 있기 때문에 순서값(인덱스)을 통해서 내부의 값에 접근할 수 있다.
- 같은 값을 중복으로 저장할 수 있다.
    - 중복되지 않은 값을 가진 순서가 있기 때문에 중복을 가려낼 수 있다.
- 요소의 값을 수정, 삭제, 추가 등을 자유롭게 할 수 있다.
- 요소가 중간에 추가되거나 삭제되면 순서값이 새롭게 부여된다.

## 리스트의 생성
- 리스트는 대괄호 `[]` 안에 저장하고자 하는 여러 값을 반점,을 이용하여 구분해 전달하면 리스트에 해당 값들이 순서대로 저장된다.
```python
my_list = [] # 비워둔 리스트 생성
my_list = [1, 2, 3, 4] # 여러 값들을 저장한 리스트 생성
my_list = [1, "홍길동", 3.14, True] # 서로 다른 타입의 값을 저장할 수 있다.
```

````python
# 리스트의 생성
# 1. 빈 리스트 생성
my_list = [] # 빈 대괄호
my_list = list()

# 2. 저장할 값을 나열하여 리스트 생성
my_list = [1, 2, 3, 4]
print(my_list)
````

````python
# 리스트 요소로의 접근
# 리스트에서 내부의 요소 값을 가져오려면 순서값을
# 대괄호[] 안에 전달해주어야 한다.
# 이때 가장 먼저 오는 순서값은 0으로 시작한다.
my_list = ["홍길동", "고길동", "둘리", "도우너"]
# 가장 먼저 들어간 "홍길동"이라는 값을 가져오고자 한다면
# my_list[0]으로 접근하여 값을 가져올 수 있다.
# my_list[0]를 우리가 기존에 하나의 값만 담던 변수라고 봐도 무방한다.
go = my_list[0]
print(go)
````

````python
# 마지막 요소로의 접근
# 방법이 두 가지 있는데, 하나는 길이를 이용한 접근법이며
# 다른 하나는 파이썬의 문법인 음수 인덱스를 이용하는 방법이다.
# 1. 길이를 추출하여 길이-1번째 요소에 접근하는 방법
names = ["홍길동", "둘리", "또치", "마이콜", "고희동"]
# 리스트의 길이 추출
length = len(names) # 리스트가 담긴 변수를 전달하면 요소의 개수를 정수로 반환한다.

# length - 1번째 요소에 접근
print(names[length-1]) # 전체 길이에서 1을 뺀 숫자값을 순서값으로 전달

# 2. -1을 전달하여 마지막 요소에 접근
# 이때 -2를 사용하면 뒤에서 두번째 요소에 접근할 수 있다.
print(names[-1]) # 마지막 요소에 접근
````

````python
# 슬라이싱(slicing)
# 리스트의 요소에서 하나의 값만 가져오는 것이 아니라
# 여러 요소를 규칙에 따라 여러 개 가져오는 방법이다.
# 이때, 규칙이라 함은 등차수열을 가리킨다.
names = ["이순신", "장보고", "장영실", "이도", "유관순"]

# 0번째 요소부터 n-1번째 요소까지 가져오고 싶으면
# [start:end] 방식으로 n을 end로 지정할 수 있다.
print(names[0:2])

# 1. names의 1번째 요소부터 3번째 요소까지 출력

# 2. 2번째부터 3번째 요소까지 새로운 리스트 new_list에 담은 뒤

# new_list의 첫번째 요소 출력
````

````python
# 슬라이싱 심화
# 대괄호[] 안에 콜론을 이용해서 여러 값을 가져올 수 있었는데,
# 만약 시작값이나 끝값을 비운 채 콜론을 사용하면
# 첫번째 혹은 마지막까지 전체를 가져오는 문법이 된다.
print(names)

# 1번째 요소부터 마지막 요소까지 가져오고 싶다.
# [1:] 뒤를 비워두면 마지막 요소까지 가져온다.
print(names[1:])

# 앞쪽의 start 부분을 비워두면 가장 처음부터 콜론 뒤의 end-1까지 가져온다.
print(names[:3])
````

````python
# 특정 요소의 값 변경하기
# 이전에 대괄호[]에 인덱스를 전달하여 특정 저장공간에 접근했다.(값을 가져옴)
# 같은 방식으로 해당 저장공간을 대입 연산자 좌항에 두면
# 해당 저장공간에 값을 새로 저장할 수 있다.
names = ['이순신', '장보고', '장영실', '이도', '유관순']
print(names)
# 3번째 요소를 "세종대왕"으로 수정, 저장
names[3] = "세종대왕"
print(names)
````

````python
# 요소의 삭제
# 변수를 삭제할 때와 같이 del 예약어를 사용하여
# 리스트에서 제거할 수 있다.
names = ['이순신', '장보고', '장영실', '이도', '유관순']
print(names)
# del 리스트명[인덱스값]
del names[0]
print(names)
print(names[0])
````

````python
# 요소의 추가
# 요소를 추가하려면 함수를 사용해야 한다.
# 리스트명.append(추가할 값)
names = ['이순신', '장보고', '장영실', '이도'] # '유관순'
print(names)
test = names.append('유관순') # 값을 추가하는 기능
# 반환값은 없다.
# print(test)
print(names)
````
